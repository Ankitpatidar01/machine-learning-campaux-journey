```python
## To get all the pages --
import time

session = requests.Session()
df_list = []

for i in range(1, 429):
    try:
        r = session.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=6fb78acbe331551f578b44114cdc9194&language=en-US&page={i}', timeout=10)
        data = r.json()['results']
        df_list.append(pd.DataFrame(data)[['id','title','overview','release_date','popularity','vote_average','vote_count']])
    except:
        time.sleep(1)
    time.sleep(0.3)

df1 = pd.concat(df_list, ignore_index=True)


    ## because each page content 1- 20 movie and each have index 1 - 20 so to continuee index from 21 
    ## second page we use ignore_index = True
    
```

```python
df1.to_csv('movie.csv')
```
