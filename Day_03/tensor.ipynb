## Basic Tensor Operation

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print("Add:\n", a+b)
print("Subtract:\n", a-b)
print("Element-wise multiply:\n", a*b)
print("Matrix multiply:\n", a @ b)
```

## Reshape & Manipulate Tensors

```python
arr = np.arange(12).reshape(3 , 4)
print('array : ' , arr)

transpose = arr.T;
print('transpose : ' , transpose)

# Adding/removing dimensions
# print('squeeze : ' , arr.squeeze().shape)

y = np.random.rand(2 , 3, 4)

print(np.expand_dims(y , axis = 0).shape)
print(y)

print('squeeze : ' , y.squeeze().shape)
print(y)
```

## Indexing and Slicing

```python
arr = np.random.rand(2 , 3 , 4)
print(arr)

print('slicing : ' , arr[:,1:,3:])
```

## Real ML Related Tasks

#### 1️⃣ One-hot encoding

```python
labels = np.array([0,2,1,2])
one_hot = np.zeros((labels.size, labels.max()+1))
one_hot[np.arange(labels.size), labels] = 1
print(one_hot)
```

#### 2️⃣ Normalization

```python
data = np.array([1 , 2, 3 , 4 , 5] , dtype = float)
normalized = ((data - data.mean())) / data.std()
print(normalized)
```

#### 3️⃣ Mini-batch creation

```python
dataset = np.arange(20).reshape(10,2)  
batch_size = 3
for i in range(0, len(dataset), batch_size):
    print("Batch:\n", dataset[i:i+batch_size])
```
